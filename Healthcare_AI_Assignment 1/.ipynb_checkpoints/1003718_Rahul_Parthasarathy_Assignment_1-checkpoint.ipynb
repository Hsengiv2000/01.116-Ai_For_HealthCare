```python
def config_gpu():
    #Configure Gpus
    gpus = tf.config.experimental.list_physical_devices('GPU')
    if gpus:
      try:
        # Currently, memory growth needs to be the same across GPUs
        for gpu in gpus:
          tf.config.experimental.set_memory_growth(gpu, True)
        logical_gpus = tf.config.experimental.list_logical_devices('GPU')
        print(len(gpus), "Physical GPUs,", len(logical_gpus), "Logical GPUs")
      except RuntimeError as e:
        # Memory growth must be set before GPUs have been initialized
        print(e)
```

```python
#Preprocessing Images Functions:
```

```python
def preprocess(name, size=128):
    image= tf.image.decode_jpeg(tf.io.read_file(name),channels=3) #load as colour image
    image= tf.image.resize(image, [128, 128], preserve_aspect_ratio=False)/255 #resize and normalize
    label=tf.strings.to_number(tf.strings.split(name, os.sep)[-2],tf.dtypes.int32) #get class name
   
    return image, label
```

```python
#Create a tensorflow pipeline and load and preprocess the data
```

```python
size = 128
DATA_DIR = 'Data'
listoffiles = tf.data.Dataset.list_files(DATA_DIR + '/*/*')
batch_size=100

data = len(listoffiles)
train_data_size = int(np.floor(0.8 * data))

preprocess_function = partial(preprocess, size=size)
processed_data = listoffiles.map(preprocess_function).shuffle(buffer_size=data,reshuffle_each_iteration=False)
train_data = processed_data.take(train_data_size).batch(batch_size)
test_data = processed_data.skip(train_data_size).batch(batch_size)
```

```python
#Checkpoint 2: Print a sample of the data and explain the selected range and the purpose of normalisation----------------------
```

```python
n=1
for image, label in train_data.take(1):
    for i in range(1):
        plt.figure()
        plt.title(label[i].numpy())
        plt.imshow(image[i])
    
```

```python
#Build Base Model (As defined in the exercise)
```

```python
inputLayer = tf.keras.Input(shape=(size, size, 3), batch_size=batch_size, dtype='float32', name='input')
x = tf.keras.layers.Conv2D(16, 3, padding='same', activation='relu')(inputLayer)
x = tf.keras.layers.MaxPooling2D()(x)
x = tf.keras.layers.Conv2D(32, 3, padding='same', activation='relu')(x)
x = tf.keras.layers.MaxPooling2D()(x)
x = tf.keras.layers.Conv2D(32, 3, padding='same', activation='relu')(x)
x = tf.keras.layers.MaxPooling2D()(x)
x = tf.keras.layers.Flatten()(x)
x = tf.keras.layers.Dense(128, activation='relu')(x)
outputLayer = tf.keras.layers.Dense(4, name='output')(x)
model = tf.keras.Model(inputs=[inputLayer], outputs=[outputLayer])
```

```python
#Checkpoint 3 Display the graph Print out of the model
print(model.summary())
```

```python
#graphfile = 'initialmodel.png'
#tf.keras.utils.plot_model(model, to_file=graphfile, show_shapes=True)
#I AM NOT ABLE TO GET THE IMPORT THE WORK SO I PRINTED THE MODEL ABOVE
```

```python
#Define Loss function and Optimiser
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

```python

#Checkpoint 4: Explain selection of loss function and optimiser Optimizer
print(" Generally a binary cross entropy loss is used for binary classification and a categorical cross entropy loss is used for multilabel classification")
print("We use Sparse Categorical Crossentropy because the each image belongs to exactly one label and not multiple labels, hence we dont use cross entropy but rather use sparse categorical crossentropy")
print("My computer in general could only support Adam and SGD as the other optimizers were very laggy, however Adam is super efficient and good at training")
print("Adam relies on the theory that learning rate is like a heavy ball and adds natural momentum to the training process, leading to faster convergence")
```

```python
#Define Training/Testing Function here
```

```python
epochs=10
params = model.fit(train_data,validation_data=test_data,epochs=epochs,)
```

```python
inputlayer = tf.keras.Input(shape=(size, size, 3), batch_size=batch_size, dtype='float32', name='input')

print('Resnet is super light and trains very fast. hence I went for resnet. It is also a great feature extractor and very good at classification. it also has lower parameters to learn. It also escapes the problem of vanishing gradient')
print("VGG is too heavy for the task and is super slow at training, hence I did not go for it")
print("Inception is another good option because of the varying kernal size that allows detection at different scales, however Resnet is lighter so I am going for resnet")

resnet = tf.keras.applications.ResNet50(include_top=False, input_tensor=inputlayer, input_shape=(size,size,3), classes=4)

chosen_model =resnet

chosen_model.trainable = False
x = chosen_model(inputlayer, training=False)
#Basically the model is untrainable now and the final layer is removed. hence we add a flatten + dense layer that acts as our classification head. 
#This dense layer is getting trained, while the convolution layers are frozen and not being trained. 
#This is because our feature extractor (Resnet backbone ) is already trained.
x = tf.keras.layers.Flatten()(x)
x = tf.keras.layers.Dense(128, activation='relu')(x)
outputlayer = tf.keras.layers.Dense(4, name='output')(x)
model = tf.keras.Model(inputs=[inputlayer], outputs=[outputlayer])
```

```python
# Preprocessing / Data Augmentation CNV Images
cnvdataset= tf.data.Dataset.list_files(DATA_DIR + '/0/*') #0 corresponds to CNV
cnvdataset = cnvdataset.map(preprocess_function) #underrepresentated class

dataset = tf.data.Dataset.list_files(DATA_DIR + '/*/*') #original dataset

procdataset = dataset.map(preprocess_function)

balanceddataset = procdataset.concatenate(cnvdataset).concatenate(cnvdataset) #sampling of underrepresentated class and appending to original dataset

datasetsize = len(balanceddataset)
balanceddataset = balanceddataset.shuffle(datasetsize, reshuffle_each_iteration=False)

balancedtrainsize = int(np.floor(0.8 * datasetsize))
balancedtraindata = balanceddataset.take(balancedtrainsize).batch(batch_size)
balancedtestdata = balanceddataset.skip(balancedtrainsize).batch(batch_size)
```

```python
inputlayer = tf.keras.Input(shape=(size, size, 3), batch_size=batch_size, dtype='float32', name='input')
x = tf.keras.layers.Conv2D(16, 3, padding='same', activation='relu')(inputlayer)
x = tf.keras.layers.MaxPooling2D()(x)
x = tf.keras.layers.Conv2D(32, 3, padding='same', activation='relu')(x)
x = tf.keras.layers.MaxPooling2D()(x)
x = tf.keras.layers.Conv2D(32, 3, padding='same', activation='relu')(x)
x = tf.keras.layers.MaxPooling2D()(x)
x = tf.keras.layers.Flatten()(x)
x = tf.keras.layers.Dense(128, activation='relu')(x)
outputlayer = tf.keras.layers.Dense(4, name='output')(x)
balancedmodel = tf.keras.Model(inputs=[inputlayer], outputs=[outputlayer]) #We are using the same vanilla model

print(balancedmodel.summary())

balancedmodel.compile(optimizer='adam',loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),metrics=['accuracy'])

epochs=10
params3 = balancedmodel.fit(balancedtraindata,validation_data=balancedtestdata,epochs=epochs)
```

```python
#Basically what I did was, i took those 1000 CNV images, made two more copies, leading to 3000 CNV images. I did not perform augmentation as the dataset is simple.
#AS you can clearly see, balancing the dataset gives a validation accuracy of 98%, which is much better than the initial performance of the model
#It is also better than the RESNET model. 
#This shows that data is gold when it comes to Computer Vision and it is very important to have a good balanced dataset
```
